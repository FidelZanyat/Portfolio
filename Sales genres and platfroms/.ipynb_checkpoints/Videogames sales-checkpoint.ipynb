# Описание данных: 
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)

# Шаг 5. Проверка гипотез
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


```python
# формируем данные для сравнения
xbox_user = data_temp[data_temp['platform'] == 'XOne']
pc_user = data_temp[data_temp['platform'] == 'PC']

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(
    xbox_user['user_score'].dropna(), 
    pc_user['user_score'].dropna(), equal_var = False)

# Формально я сравниваю Xbox One, но для удобства буду писать просто Хбокс / Xbox для краткости. 
print('среднее оценок пользователей Хбокса', xbox_user['user_score'].mean())
print('среднее оценок пользователей ПК', pc_user['user_score'].mean())
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Мы проверяем гипотезу, что оценки пользователей ПК и Хбокс одинаковые. Отвергнуть её не вышло. Возможно вкусы пользователей совпадают. 

```python
action_user = data_temp[data_temp['genre'] == 'Action']
sport_user = data_temp[data_temp['genre'] == 'Sports']

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(
    action_user['user_score'].dropna(), 
    sport_user['user_score'].dropna(), equal_var = False)

print('среднее оценок Action`ов', action_user['user_score'].mean())
print('среднее оценок спортвных игр', sport_user['user_score'].mean())
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

А тут мы проверяем, одинаковые ли оценки у спортсимов и экшенов. Не одинаковые, отбрасываем нулевую и подтверждаем, что да, оценки разные.
